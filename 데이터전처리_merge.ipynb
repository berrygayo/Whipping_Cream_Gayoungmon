## 티어 점수화 
> 브론즈 300점 시작 단계별 25점

```python
# 티어 점수 표 생성 
tiar = ['Iron 4', 'Iron 3','Iron 2', 'Iron 1',
        'Bronze 4', 'Bronze 3', 'Bronze 2', 'Bronze 1',
        'Silver 4', 'Silver 3', 'Silver 2', 'Silver 1',
        'Gold 4', 'Gold 3', 'Gold 2', 'Gold 1',
        'Platinum 4', 'Platinum 3', 'Platinum 2', 'Platinum 1',
        'Diamond 4', 'Diamond 3', 'Diamond 2', 'Diamond 1',
        'Master','Grand Master', 'Challenger'
       ]
```

```python
tiar_score = []
for i in range(len(tiar)):
    tiar_score.append(200 + 25*i)
tiar_scrore_df=pd.DataFrame({"TIAR":tiar, "TIAR_SCORE":tiar_score})
tiar_scrore_df
```

## '승','패' 제거 후 정수화 

```python
# '승','패' 한글 제거 
def del_str(a):
    a=a.replace('승','')
    a=a.replace('패','')
    a=int(a)
    return a 

#del_str('3승패')
```

```python
for i in range(len(test_data)):
    test_data['TEAM_WIns'][i] = del_str(test_data['TEAM_WIns'][i])
    test_data['SOLO_WIns'][i] = del_str(test_data['SOLO_WIns'][i])
    test_data['TEAM_Losses'][i] = del_str(test_data['TEAM_Losses'][i])
    test_data['SOLO_Losses'][i] = del_str(test_data['SOLO_Losses'][i])
```

```python
test_data.info()
```

```python
# object > int 변경 
test_data = test_data.astype({'TEAM_WIns': 'int','SOLO_WIns': 'int','TEAM_Losses': 'int','SOLO_Losses': 'int' })
```
